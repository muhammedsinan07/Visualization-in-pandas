```python
categories = ['category1','category2','category3']
values = [40,54,20]

colors=['red','yellow','blue']
explode=(0,0.2,0)

plt.pie(values,labels=categories,autopct='%1.2f%%',colors=colors,explode=explode,radius=2,shadow=True)# autopct-shows percentage

plt.title('Sample pie chart')
plt.show()
```

- cant use directely the argument 'edgecolor' in plt.pie
- we can use 'wedgeprops' argument

```python
categories = ['category1','category2','category3']
values = [40,54,20]

colors=['red','yellow','blue']
explode=(0,0.2,0)
wedgeprops={'edgecolor':'k','linewidth':3}

plt.pie(values,labels=categories,autopct='%1.2f%%',colors=colors,explode=explode,wedgeprops=wedgeprops)

plt.title('Sample pie chart')
plt.show()
```

### Subplots

```python
x = np.linspace(0,8,400)
y1 = np.sin(x)
y2 = np.sin(x**2)
y3 = np.sin(x**3)
y4 = y1**2
y5 = y2**2
y6 = y3**2
```

```python
a,b = plt.subplots(nrows=2,ncols=3)

b[0,0].plot(x,y1,color='red')
b[0,1].plot(x,y2,color='red')
b[0,2].plot(x,y3,color='red')
b[1,0].plot(x,y4,color='blue')
b[1,1].plot(x,y5,color='blue')
b[1,2].plot(x,y6,color='blue')

#title
b[0,0].set_title('Plot with sin(x)')
b[0,1].set_title('Plot with sin(x**2)')
b[0,2].set_title('Plot with sin(x**3)')
b[1,0].set_title('Plot with y1**2')
b[1,1].set_title('Plot with y2**2')
b[1,2].set_title('Plot with y3**2')
plt.suptitle('Stacked Plots')

plt.show()
```
